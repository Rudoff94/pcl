```python
import numpy as np
import pandas as pd
import sklearn
from matplotlib import pyplot as plt
import matplotlib
import cv2
import os
matplotlib.rcParams['figure.figsize'] = 5, 10
```

```python
photo_dir_path = os.path.expanduser('~') + '/photo/511p/'
photo_name = '511-23-50.jpg'
source_img = cv2.imread(photo_dir_path + photo_name)
source_img = cv2.cvtColor(source_img, cv2.COLOR_BGR2RGB)
```

```python
hsv = cv2.cvtColor(source_img, cv2.COLOR_RGB2HSV)
plt.imshow(source_img)
```

```python
gray = cv2.cvtColor(source_img, cv2.COLOR_RGB2GRAY)
filtered_ln10 = np.log10(gray).astype(np.uint8)
filtered_ln2 = np.log2(gray).astype(np.uint8)
filtered_ln = np.log(gray).astype(np.uint8)
filtered_sqrt = np.sqrt(gray).astype(np.uint8)

#corrected_rgb = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
#plt.imshow(corrected_rgb)

filtered_ln10 = filtered_ln10.flatten()
filtered_ln2 = filtered_ln2.flatten()
filtered_ln = filtered_ln.flatten()
filtered_sqrt = filtered_sqrt.flatten()
source = gray.flatten()

ig, ax = plt.subplots(1,5,figsize=(20,10))
hist_value = 20
ax[0].hist(source, hist_value)
ax[0].set_title('origin')

ax[1].hist(filtered_ln10, hist_value)
ax[1].set_title('filtered_ln10')

ax[2].hist(filtered_ln2, hist_value)
ax[2].set_title('filtered_ln2')

ax[3].hist(filtered_ln, hist_value)
ax[3].set_title('filtered_ln')

ax[4].hist(filtered_ln, hist_value)
ax[4].set_title('sqrt')
```

```python
filtered_gray = np.log2(gray).astype(np.uint8)
#filtered_gray = filtered_gray * 1024

ig, ax = plt.subplots(1,6,figsize=(30,15))
ax[0].imshow(filtered_gray == 3)
ax[0].set_title('class_3')

ax[1].imshow(filtered_gray == 4)
ax[1].set_title('class_4')

ax[2].imshow(filtered_gray == 5)
ax[2].set_title('class_5')

ax[3].imshow(filtered_gray == 6)
ax[3].set_title('class_6')

ax[4].imshow(filtered_gray == 7)
ax[4].set_title('class_7')

ax[5].imshow(filtered_gray)
ax[5].set_title('origrn')
```

```python
filtered_gray_check = filtered_gray.copy()
filtered_gray_check[filtered_gray_check != 6] = 0 
filtered_gray_check = cv2.GaussianBlur(filtered_gray_check,(21, 21),0)
laplacian = cv2.Laplacian(filtered_gray_check,cv2.CV_8U)
#edges = cv2.Canny(filtered_gray_check, 100, 200)
plt.imshow(laplacian)
```
