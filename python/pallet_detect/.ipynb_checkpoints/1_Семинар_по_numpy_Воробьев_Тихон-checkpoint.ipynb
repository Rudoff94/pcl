```python
my_np_array.astype(np.complex)
```

```python
s = a[4:7]
s[:] = 10
a
```

Пример

```python
import matplotlib.pylab as plt
%matplotlib inline
import itertools
```

```python
import time
%time
plt.figure(figsize=(10, 10))
lena = plt.imread('lena.jpg')
permutation = list(itertools.permutations([0,1,2]))
for i in range(6):
    plt.subplot(2, 3, i+1)
    temp_lena = list(map(lambda x: x[:,permutation[i]], lena))
    plt.imshow(temp_lena)
```
